```python

import pandas as pd
df = pd.read_csv('data.csv', delimiter = ',', decimal = ".", header = 0)
```

```python
df.head()
```

```python
df['Experiencia con el uso de la autenticación biométrica'].value_counts().plot(kind='pie', fontsize=20, autopct='%.2f%%',  title='Experiencia con el uso de la autenticación biométrica',  ylabel='', figsize=(12.5, 12.5))
```

```python
df['Tipo de autenticación biometrica más utilizado'].value_counts().plot(kind='pie', fontsize=20, autopct='%.2f%%',  title='Tipo de autenticación biometrica más utilizado',  ylabel='', figsize=(12.5, 12.5))
```

```python
plt = df['Tipo de autenticación más frecuente'].value_counts().plot(kind='pie', fontsize=20, autopct='%.2f%%',  title='Tipo de autenticación más frecuente',  ylabel='', figsize=(12.5, 12.5))
```

```python
plt = df['Perspectiva sobre la vulneración de la privacidad con el uso biométrico'].value_counts().plot(kind='barh', fontsize=20,  title='Perspectiva sobre la vulneración de la privacidad con el uso biométrico',  ylabel='', figsize=(12.5, 12.5))   
```
